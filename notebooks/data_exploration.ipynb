# Cleaning the data

```python
# Handle missing values by dropping rows with any missing data
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()

# Standardize categorical values 
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    df[col] = df[col].str.strip().str.lower()

# Filter rows based on specific conditions
if 'sales' in df.columns:
    df = df[df['sales'] >= 0]

df.head()
```

# Data Visualizations

```python
# Heatmap of missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()
```

```python
# Distribution of numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Revenue Generated vs Budget
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Budget', y='Revenue_Generated', data=data, hue='Subscription_Tier', palette='viridis')
plt.title('Revenue Generated vs Budget', fontsize=16)
plt.xlabel('Budget', fontsize=12)
plt.ylabel('Revenue Generated', fontsize=12)
plt.legend(title='Subscription Tier')
plt.show()
```

```python
# Average Customer Satisfaction by Common Keywords
plt.figure(figsize=(12, 6))
sns.barplot(x='Common_Keywords', y='Customer_Satisfaction_Post_Refund', data=data, ci=None, palette='coolwarm')
plt.title('Average Customer Satisfaction by Common Keywords', fontsize=16)
plt.xlabel('Common Keywords', fontsize=12)
plt.ylabel('Average Customer Satisfaction', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Units Sold by Discount Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='Discount_Level', y='Units_Sold', data=data, palette='Set2')
plt.title('Units Sold by Discount Level', fontsize=16)
plt.xlabel('Discount Level', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.show()
```
